```python
import geopandas as gpd
import matplotlib.pyplot as plt
from pyproj import CRS
```

```python
admin = gpd.read_file("Admin.geojson")
tissot = gpd.read_file("Points_Buffer.geojson")
```

```python
def tissot_map(epsg):
  fig, ax = plt.subplots(dpi=300)
  admin.to_crs(epsg) \
    .plot(ax=ax,
          color="grey",
          zorder=5)
  tissot.to_crs(epsg) \
    .plot(ax=ax, 
          color="#1f78b4",
          zorder=10)
  ax.set_title(CRS.from_epsg(epsg).name)
  ax.set_aspect("equal")
```

```python
tissot_map(4326)
```

```python
tissot_map(3395)
```
